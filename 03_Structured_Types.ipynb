# Demo 1

A rural health clinic stores basic patient data. We will use a list of tuples, whereby each tuple holds one patient's details, to practise list creation, indexing, slicing, appending, and built-in functions.

```python
patient_records = [
    (101, 35, "Female"),
    (102, 68, "Male"),
    (103, 50, "Male"),
]

print("="*30)
print("PATIENT RECORDS")
print("="*30)
# accessing the element
for id, age, gender in patient_records:
  print(f"{id}\t{age}\t{gender}")

#for patient in patient_records:
#  print(f"{patient[0]}\t{patient[1]}\t{patient[2]}")

#for patient in patient_records:
#  id, age, gender = patient
#  print(f"{id}\t{age}\t{gender}")
```

```python
print("="*30)
print("FIRST PATIENT RECORDS")
print("="*30)
# accessing specific element
print(f"{patient_records[0][0]}\t{patient_records[0][1]}\t{patient_records[0][2]}")

print("="*30)
print("LAST PATIENT RECORDS")
print("="*30)
# accessing specific element
print(f"{patient_records[-1][0]}\t{patient_records[-1][1]}\t{patient_records[-1][2]}")
```

```python
# updating patient records
patient_records.append((104, 40, "Female"))
patient_records.append((105, 55, "Male"))

print("="*30)
print("PATIENT RECORDS")
print("="*30)
# accessing the element
for id, age, gender in patient_records:
  print(f"{id}\t{age}\t{gender}")
```

```python
ages = [patient[1] for patient in patient_records]
print(f"Total number of patient: {len(patient_records)}")
print(f"Oldest patient: {max(ages)}")
print(f"Youngest patient: {min(ages)}")
```

# Demo 2

An emergency response team tracks resource quantities at a depot using a dictionary. We practise creating, reading, updating, and deleting key-value pairs, and iterating with .items(), .keys(), .values().

```python
# declaration
depot_stock = {
    "Bandages": 100,
    "Water Bottles": 500,
    "First Aid Kits": 200,
}

print("="*30)
print("INVENTORY")
print("="*30)
for item, qty in depot_stock.items():
  print(f"{item}\t\t{qty}")

# update values
depot_stock["Bandages"] += 150

# add new element
depot_stock["Blankets"] = 300
print("After Adding New Key:")
for item, qty in depot_stock.items():
  print(f"{item}\t\t{qty}")

# delete a key
del depot_stock["Water Bottles"]
print("After Deleting a Key")
for item, qty in depot_stock.items():
  print(f"{item}\t\t{qty}")
```

```python
total_items = 0
depot_stock["Maggi"] = 50

print("="*30)
print("INVENTORY")
print("="*30)
for item, qty in depot_stock.items():
  total_items += qty
  if qty < 200:
    print(f"{item:<20}{qty}\t LOW STOCK")
  else:
    print(f"{item:<20}{qty}\t")

print(f"Total number of items:  {total_items}")
```

# Demo 3

Central Park (15,000 sqm)
    Amenities: Playground, Fountain
  Riverside Gardens (8,000 sqm)
    Amenities: Walking Trail
  Maple Grove (5,000 sqm)
    Amenities: Picnic Tables, BBQ Area, Playground


```python
parks = [
    {"name": "Central Park",
     "size": 15000,
     "amenities": ["Playground", "Fountain"]},
    {"name": "Riverside Gardens",
     "size": 8000,
     "amenities": ["Walking Trail"]},
    {"name": "Maple Grove",
     "size": 5000,
     "amenities": ["Picnic Tables", "BBQ Area", "Playground"]},
]

sizes = [p["size"] for p in parks]
largest_park = sizes.index(max(sizes))
print(f"Largest park is {parks[largest_park]["name"]} with {max(sizes)} sqm")

print("The following park has playground:")
for p in parks:
  if "Playground" in p["amenities"]:
    print(f"  {p["name"]}")

print("="*30)
print("PARK DIRECTORY")
print("="*30)
for p in parks:
  print(f"{p["name"]:<20} {p["size"]} sqm \t")
  for amenity in p["amenities"]:
    print(f"  {amenity}")
print(f"Total parks now: {len(parks)}")
```
