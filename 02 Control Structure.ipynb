# Demo 3: Weekly Plastic Waste Collection Tracker

A recycling centre records daily plastic waste collected (kg) over one week. The daily target is 50 kg. We will use a for loop to process each day’s data: [45.5, 52.0, 38.7, 61.2, 49.0, 55.1, 40.8] and report whether the target was met.

```python
target = 50

daily_data = [45.5, 52.0, 38.7, 61.2, 49.0, 55.1, 40.8]

days_met = 0
days_missed = 0
total_kg = 0

print("="*50)
print("    WEEKLY PLASTIC WASTE COLLECTION REPORT")
print("="*50)

for day, waste in enumerate(daily_data, start=1):
    if waste >= target:
        print(f"Day {day}: Target met!   Collected {waste:.2f} kg.")
        days_met += 1
    else:
        print(f"Day {day}: Below target! Collected {waste:.2f} kg only.")
        days_missed += 1
    total_kg += waste

average_kg = total_kg / len(daily_data)

print("="*50)
print(f"Days target met  : {days_met}")
print(f"Days below target: {days_missed}")
print(f"Total Waste      : {total_kg}")
print(f"Average Waste    : {average_kg}")
print("="*50)
```

# Demo 4: Customer Feedback Rating Collector

A university cafeteria collects customer satisfaction ratings (1–5). The system keeps collecting until the staff enters 0 to stop. We will use a while loop with a match-case inside to count each rating category.

Example run (input: 5, 3, 1, 5, 0)


```python
excellent = good = average = poor = very_poor = 0
total_rating = 0

print("=== Cafeteria Feedback System === ")
print("Enter ratings (1-5). Enter 0 when done.")
print("-"*50)

rating = int(input("Rating: "))

while rating != 0:
  match rating:
    case 5:
      print(" → Recorded Excellent")
      excellent += 1
    case 4:
      print(" → Recorded Good")
      good += 1
    case 3:
      print(" → Recorded Average")
      average += 1
    case 2:
      print(" → Recorded Poor")
      poor += 1
    case 1:
      print(" → Recorded Very Poor")
      very_poor += 1

  total_rating += 1
  rating = int(input("Rating: "))

if total_rating == 0:
  satisfaction_score = 0
else:
  satisfaction_score = (5*excellent + 4*good + 3*average + 2*poor + 1*very_poor) / total_rating

print("-"*50)
print(f"Total ratings received: {total_rating}")
print(f"   Excellent (5)  :{excellent}")
print(f"   Good      (4)  :{good}")
print(f"   Average   (3)  :{average}")
print(f"   Poor      (2)  :{poor}")
print(f"   Very Poor (1)  :{very_poor}")
print(f"Satisfaction Score:{satisfaction_score}")
```
